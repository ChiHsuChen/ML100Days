```
import pandas
import numpy as np
import os

df_app_train=pandas.read_csv('./application_train.csv')
df_app_train.head()
```

```
df_WEEKDAY_APPR_PROCESS_START=pandas.DataFrame(df_app_train['WEEKDAY_APPR_PROCESS_START'])
df_WEEKDAY_APPR_PROCESS_START.head()
```

```
# label encoding
from sklearn.preprocessing import LabelEncoder

lbencoder=LabelEncoder()
lblist=lbencoder.fit_transform(df_WEEKDAY_APPR_PROCESS_START)
df_lblist=pandas.DataFrame(lblist)
df_lblist.head()
```

```
# one hot encoding
from sklearn.preprocessing import OneHotEncoder

# 先規範一周
onehotencoder=OneHotEncoder()

# 使用one hot encoder轉換
onehotlist=onehotencoder.fit_transform(df_WEEKDAY_APPR_PROCESS_START).toarray()
df_onehotlist=pandas.DataFrame(onehotlist)
df_onehotlist.head()
```
