```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# 弱相關

```
# X,Y隨機初始化,1-500
X=np.random.randint(1,500,1000)
Y=np.random.randint(1,500,1000)

# 求cofficient
np.corrcoef(X,Y)
```

```
# 畫出scatter plot
plt.scatter(X,Y)
```

# 正相關

```
# X,Y隨機初始化,1-500
X=np.random.randint(1,500,1000)
Y=X+np.random.randint(1,50,1000)  # 讓y->x再加上一些隨機數後，變正相關

# 求cofficient
np.corrcoef(X,Y)
```

```
# 畫出scatter plot
plt.scatter(X,Y)
```

# 負相關

```
# X,Y隨機初始化,1-500
X=np.random.randint(1,500,1000)
Y=-X+np.random.randint(1,50,1000)  # 讓y->-x再加上一些隨機數後，變負相關

# 求cofficient
np.corrcoef(X,Y)
```

```
# 畫出scatter plot
plt.scatter(X,Y)
```
