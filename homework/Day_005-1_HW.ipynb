在小量的資料上，我們用眼睛就可以看得出來程式碼是否有跑出我們理想中的結果。請嘗試想像一個你需要的資料結構 (裡面的值是隨機的)，然後用程式碼範例的方法把它變成 DataFrame

```
# 方法一
import pandas

sample_data={
    'env':['SYS1','SYS2','SYS3'],
    'avgtime':[0.1,0.123,0.342]
}

df_sampledata=pandas.DataFrame(sample_data)
df_sampledata.head()
```

```
# 方法2
env=['SYS1','SYS2','SYS3']
avgtime=[0.1,0.123,0.342]

listlabels=['env','avgtime']
listcols=[env,avgtime]

zipdata=list(zip(listlabels,listcols))

df_sampledata2=pandas.DataFrame(dict(zipdata))
df_sampledata2.head()
```

```
# 依據範例求出group by的mean
df_sampledata2.groupby(by='env')['avgtime'].mean()
```
