# 思考題_請找出一個你認為機器學習可以發光發熱的領域並闡述原因?

1. 領域：
* L5自動駕駛
2. 原因：
* 應該不是單純從"人不用踩油門、握方向盤"來看待自動駕駛，而是思考L5自動駕駛實現後會有那些經濟模式被改寫
* 比如：無人自助商店就可以改為裝載於貨櫃，由無人拖車頭自動駕駛到系統預判購買熱點，從本來是定點式改為不定點行動化
* 而本來車輛共享服務也成為可能，一個家庭周末時間之外的每日用車時間比例極小，但購入新車的折舊卻是每年固定15%
試想如果有車可以自動每天在上班時間定點到家接送，還可以在自訂中途繞道小孩學校，而下班時還可以可能加班造成下班delay還可以自行調整讓車子先到學校接小孩回家，而不用自己開車接小孩時還得跟時間賽跑；住在蛋黃區的民眾的購車策略將因此改變，對這些人就會免去新車折舊、保險費、維護費以及自家車事故帶來的不便；而這些人本來就是比較有能力負擔高單價車款的階層，進而造成車廠的新車策略也跟著大幅調整
