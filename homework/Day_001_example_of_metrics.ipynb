請點選下方檢視範例依照 Day_001_example_of_metrics.ipynb 完成 Mean Squared Error 的函式。

```
import numpy as np

def meansquareerror(A,B):
  result=np.mean((A-B)**2)
  # 或 result=((A-B)**2)/len(A)

  return result

testA=np.array([1,2,3])
testB=np.array([2,5,7])

print('arrar1{} array2{} mean square error={}'.format(testA,testB,meansquareerror(testA,testB)))
```
